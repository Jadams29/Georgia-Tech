# Finding optimal weights for Neural Networks
## Back propagation versus Ramdomized optimization

## 1. Objective: 
The purpose of this project is to find the optimal weights for Neural Network using Randomized Optimization and compare the perfromance against the perfromance of Neural Network with back propagation.

## 2. Datasets

### Wine Dataset
UCI Wine Quality dataset consists physicochemical and sensory data red and white variants of the Portuguese "Vinho Verde" wine.  Dataset is available at https://archive.ics.uci.edu/ml/datasets/Wine+Quality

Number of Instances: red wine - 1599  
Number of Attributes: 11 + output attribute

Attribute Information
Input variables (based on physicochemical tests):
1 - fixed acidity
2 - volatile acidity
3 - citric acid
4 - residual sugar
5 - chlorides
6 - free sulfur dioxide
7 - total sulfur dioxide
8 - density
9 - pH
10 - sulphates
11 - alcohol  
Output variable (based on sensory data):
12 - quality (score between 0 and 10)

## 3. Data Preprocessing

```python
#Packages Used
import pandas as pd
import numpy as np
import utils
import time
import mlrose_hiive as mlrose
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import StratifiedKFold, train_test_split, cross_val_predict, cross_val_score, GridSearchCV
from sklearn.neural_network import MLPClassifier


from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report 
from sklearn.metrics import average_precision_score, precision_recall_curve
from sklearn.metrics import plot_precision_recall_curve

from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
%matplotlib inline
```

```python
#load red wine dataset
dataset_loc_red = "./datasets/wine/winequality-red.csv"
df_red = pd.read_csv(dataset_loc_red, sep=";")

df_red.head()
```

```python
# Categorize wine into "above average" and "below average" wines
df_red["quality"].values[df_red["quality"] <= 5] = 0
df_red["quality"].values[df_red["quality"] > 5] = 1
```

```python
Wine_X = df_red.iloc[:,:-1]
Wine_y = df_red.iloc[:,-1]
```

```python
# Split the data into a training set and a test set
Wine_X_train, Wine_X_test, Wine_y_train, Wine_y_test = train_test_split(Wine_X, Wine_y, random_state=1, test_size=0.30)
```

```python
scaler = StandardScaler()
Wine_X_train = scaler.fit_transform(Wine_X_train)
Wine_X_test = scaler.transform(Wine_X_test)
model_stats=[]
cv=10
```

## 4. Neural Network with Back Propagation 

### 4.1 Default Parameters

```python
Wine_mlpclf =  MLPClassifier(random_state=1)

Wine_y_pred = cross_val_predict(Wine_mlpclf, Wine_X_train, Wine_y_train, cv=cv, n_jobs=-1)

cnf_matrix = confusion_matrix(Wine_y_train, Wine_y_pred)
np.set_printoptions(precision=2)

# Plot non-normalized confusion matrix
plt.figure()
utils.plot_confusion_matrix(cnf_matrix, classes=[0,1],
                      title='MLP (Default) - Wine')
```

```python
modelid = "MLP (Default) - Wine"
prscore = precision_score(Wine_y_train, Wine_y_pred, average='binary')
rcscore = recall_score(Wine_y_train, Wine_y_pred, average='binary')
f1score = f1_score(Wine_y_train, Wine_y_pred, average='binary')

model_stats.append([modelid, prscore, rcscore, f1score])
print("Precision: "+str("{:.2f}".format(prscore))+", Recall: "+str("{:.2f}".format(rcscore))+", F1 Score: "+str("{:.2f}".format(f1score)))
```

```python
#plot learning curve
utils.plot_learning_curve(Wine_mlpclf, "MLP (Default) - Wine - Learning Curve", Wine_X_train, Wine_y_train, cv=cv, n_jobs=-1, exp_score=0.8, train_sizes=np.linspace(0.05, 1.0, 10), ylim=[0.5,1])
```

### 4.2 Hyperparameter Tuning - Iteration 1

```python
h_units = [10, 20, 30, 40, 50, 60, 70, 80, 90]
learning_rates = np.arange(0.0, 0.005, 0.0005)
param_grid = {'hidden_layer_sizes': h_units, 'learning_rate_init': learning_rates}

Wine_mlpgs= GridSearchCV(Wine_mlpclf, param_grid, cv=cv, scoring='f1', n_jobs=-1)
Wine_mlpgs.fit(Wine_X_train,Wine_y_train)
```

```python
print("Best Score:" + str(Wine_mlpgs.best_score_))
print("Best Parameters: " + str(Wine_mlpgs.best_params_))
print("Best Estimator: " + str(Wine_mlpgs.best_estimator_))
```

```python
Wine_mlpclf =  MLPClassifier(hidden_layer_sizes=Wine_mlpgs.best_params_["hidden_layer_sizes"], learning_rate_init=Wine_mlpgs.best_params_["learning_rate_init"], random_state=1)

Wine_y_pred = cross_val_predict(Wine_mlpclf, Wine_X_train, Wine_y_train, cv=cv, n_jobs=-1)

cnf_matrix = confusion_matrix(Wine_y_train, Wine_y_pred)
np.set_printoptions(precision=2)

# Plot non-normalized confusion matrix
plt.figure()
utils.plot_confusion_matrix(cnf_matrix, classes=[0,1],
                      title='MLP (Itr 1) - Wine')
```

```python
modelid = "MLP (Itr 1) - Wine"
prscore = precision_score(Wine_y_train, Wine_y_pred, average='binary')
rcscore = recall_score(Wine_y_train, Wine_y_pred, average='binary')
f1score = f1_score(Wine_y_train, Wine_y_pred, average='binary')

model_stats.append([modelid, prscore, rcscore, f1score])
print("Precision: "+str("{:.2f}".format(prscore))+", Recall: "+str("{:.2f}".format(rcscore))+", F1 Score: "+str("{:.2f}".format(f1score)))
```

```python
#plot learning curve
utils.plot_learning_curve(Wine_mlpclf, "MLP (Itr 1) - Wine - Learning Curve", Wine_X_train, Wine_y_train, cv=cv, n_jobs=-1, exp_score=0.8, train_sizes=np.linspace(0.05, 1.0, 10), ylim=[0.5,1])
```

### 4.3 Hyperparameter Tuning - Iteration 2

```python
parameter_range =['identity', 'logistic', 'tanh', 'relu']
param_name = "activation"

utils.plot_validation_curve(MLPClassifier(), Wine_X_train, Wine_y_train, param_name, parameter_range, cv=cv, n_jobs=-1,
                      title="MLP (Itr 2) - Wine - Validation Curve - activation", xlabel=param_name, ylim= (0.1, 1.1))
```

```python
Wine_mlpclf =  MLPClassifier(hidden_layer_sizes=30, learning_rate_init=0.0035, activation='relu', max_iter=5, random_state=1)

Wine_y_pred = cross_val_predict(Wine_mlpclf, Wine_X_train, Wine_y_train, cv=cv, n_jobs=-1)

cnf_matrix = confusion_matrix(Wine_y_train, Wine_y_pred)
np.set_printoptions(precision=2)

# Plot non-normalized confusion matrix
plt.figure()
utils.plot_confusion_matrix(cnf_matrix, classes=[0,1],
                      title='MLP (Itr 2) - Wine')
```

```python
modelid = "MLP (Itr 2) - Wine"
prscore = precision_score(Wine_y_train, Wine_y_pred, average='binary')
rcscore = recall_score(Wine_y_train, Wine_y_pred, average='binary')
f1score = f1_score(Wine_y_train, Wine_y_pred, average='binary')

model_stats.append([modelid, prscore, rcscore, f1score])
print("Precision: "+str("{:.2f}".format(prscore))+", Recall: "+str("{:.2f}".format(rcscore))+", F1 Score: "+str("{:.2f}".format(f1score)))
```

```python
#plot learning curve
utils.plot_learning_curve(Wine_mlpclf, "MLP (Itr 2) - Wine - Learning Curve", Wine_X_train, Wine_y_train, cv=cv, n_jobs=-1, exp_score=0.8, train_sizes=np.linspace(0.05, 1.0, 10), ylim=[0.5,1])
```

### 4.4 Test Set Validation

```python
t_start = time.time()*1000.0
Wine_mlpclf = Wine_mlpclf.fit(Wine_X_train, Wine_y_train)
t_stop = time.time()*1000.0
traintime = t_stop - t_start

t_start = time.time()*1000.0
Wine_y_pred = (Wine_mlpclf.predict_proba(Wine_X_test)[:,1] >= 0.4).astype(bool)
t_stop = time.time()*1000.0
testtime = t_stop - t_start

cnf_matrix = confusion_matrix(Wine_y_test, Wine_y_pred)
np.set_printoptions(precision=2)

# Plot non-normalized confusion matrix
plt.figure()
utils.plot_confusion_matrix(cnf_matrix, classes=[0,1], title='MLP (Test) - Wine')
 
modelid = "MLP (Test) - Wine"
prscore = precision_score(Wine_y_test, Wine_y_pred, average='binary')
rcscore = recall_score(Wine_y_test, Wine_y_pred, average='binary')
f1score = f1_score(Wine_y_test, Wine_y_pred, average='binary')

model_stats.append([modelid, prscore, rcscore, f1score, traintime, testtime])
print("Precision: "+str("{:.2f}".format(prscore))+", Recall: "+str("{:.2f}".format(rcscore))+", F1 Score: "+str("{:.2f}".format(f1score))+", Train Time: "+str("{:.2f}".format(traintime))+", Test Time: "+str("{:.2f}".format(testtime)))
```

**Classifier with default sklearn parameters:** Learning curve shows high training & cv score and small gap between training and cv curves which indicates that model has low variance and low bias.   
**Hyperparameter Tuning – Iteration 1:**   In this iteration optimal values of hidden units and learning rates were searched using GridSearchCV. Optimal value for hidden layers was 30 and learning rate was 0.0035. Performance of the model  improved slightly using these hyperparameters. Model still has low variance and low bias.   
**Hyperparameter Tuning – Iteration 2:** As activation function is crucial to represent non-linear complex functions, in this step step I carried out model complexity on activation function. Relu activation function gave better results which is also evident from the validation curves plotted for each of the datasets.  
Learning curves plotted during this iteration shows that for wine dataset training and cv curves merge. F1 score remained the same.  
**Performance of Test Set:** Performance of tuned model on the held out test dataset was up to the expected value. F1 score on test data was 0.72

## 5 Random Hill Climbing

### 5.1 Model Tuning using Randomized Optimization Parameters

```python
model_stats = []
for mi in [100, 500, 1000]:
    for ma in [10, 50, 100]:
        for lr in [0.001, 0.01, 1]:
            
            nn_rhc_clf = mlrose.NeuralNetwork(hidden_nodes = [30], activation = 'relu',
                                         algorithm = 'random_hill_climb', max_iters = mi,
                                         bias = True, is_classifier = True, learning_rate = lr, 
                                         max_attempts= ma, random_state = 11)

            t_start = time.time()*1000.0
            nn_rhc_clf = nn_rhc_clf.fit(Wine_X_train, Wine_y_train)
            t_stop = time.time()*1000.0
            traintime = t_stop - t_start

            t_start = time.time()*1000.0
            Wine_y_pred = cross_val_predict(nn_rhc_clf, Wine_X_train, Wine_y_train, cv=cv, n_jobs=-1)
            t_stop = time.time()*1000.0
            testtime = t_stop - t_start

            prscore = precision_score(Wine_y_train, Wine_y_pred, average='binary')
            rcscore = recall_score(Wine_y_train, Wine_y_pred, average='binary')
            f1score = f1_score(Wine_y_train, Wine_y_pred, average='binary')

            # MI=100, MA=10, LR=0.001
            model_stats.append(['RHC', mi, ma, lr, "{:.2f}".format(prscore), "{:.2f}".format(rcscore), "{:.2f}".format(f1score), traintime, testtime])
```

```python
dfstats_RHC = pd.DataFrame(model_stats, columns=['Model', 'Max_Iterations', 'Max_Attempts', 'Learn_rate', 'PRScore', 'RCScore', 'F1Score', 'TrainTime', 'TestTime'])
dfstats_RHC.sort_values(['F1Score', 'TestTime'], ascending=[False, True]).iloc[0]
```

```python
dfstats_RHC[['PRScore','RCScore','F1Score']]=dfstats_RHC[['PRScore','RCScore','F1Score']].astype(float)
fig, ax = plt.subplots(figsize=(15,7))
dfstats_RHC.groupby(['Max_Iterations','Max_Attempts','Learn_rate']).max()['F1Score'].unstack().plot(ax=ax)

plt.title('RHC: Fitness Vs Max_iter, Max_Attempts & Learn_rate')
plt.legend(loc='lower right', title="Learning Rate")
plt.xlabel('(Max Attempts, Max Iterations)')
plt.ylabel('Fitness')
plt.minorticks_on()
plt.grid(b=True, which='major', color='k', linestyle='-', alpha=0.1)
plt.grid(b=True, which='minor', color='r', linestyle='-', alpha=0.05)
plt.show()
```

```python
# Random Hill Climbing - Restart

model_stats = []
for rstrt in (0, 1, 2, 3, 4, 5):
    
    nn_rhc_clf = mlrose.NeuralNetwork(hidden_nodes = [30], activation = 'relu',
                                         algorithm = 'random_hill_climb', max_iters = 500,
                                         bias = True, is_classifier = True, learning_rate = 1, 
                                         max_attempts= 50, restarts=rstrt, random_state = 11)

    t_start = time.time()*1000.0
    nn_rhc_clf = nn_rhc_clf.fit(Wine_X_train, Wine_y_train)
    t_stop = time.time()*1000.0
    traintime = t_stop - t_start

    t_start = time.time()*1000.0
    Wine_y_pred = cross_val_predict(nn_rhc_clf, Wine_X_train, Wine_y_train, cv=cv, n_jobs=-1)
    t_stop = time.time()*1000.0
    testtime = t_stop - t_start

    prscore = precision_score(Wine_y_train, Wine_y_pred, average='binary')
    rcscore = recall_score(Wine_y_train, Wine_y_pred, average='binary')
    f1score = f1_score(Wine_y_train, Wine_y_pred, average='binary')

    # MI=100, MA=10, LR=0.001
    model_stats.append(['RHC', rstrt, "{:.2f}".format(prscore), "{:.2f}".format(rcscore), "{:.2f}".format(f1score), traintime, testtime])
```

```python
dfstats_RHC_Rstart = pd.DataFrame(model_stats, columns=['Model', 'Restarts', 'PRScore', 'RCScore', 'F1Score', 'TrainTime', 'TestTime'])
dfstats_RHC_Rstart.sort_values(['F1Score', 'TestTime'], ascending=[False, True]).iloc[0]
```

**Model Tuning:** Hidden nodes and activation function were two hyperparameters of the Neural Network tuned using backpropagation method. These parameters were frozen to optimal values of 30 and 'relu'. Parameters for Random Hill Climbing algorithm were tuned during this step. 
Optimal values of Max Attempts, Max Iterations and Learning rate were 50, 1000 and 1 respectively.
Tuning of Restart parameter was carried out and its optimal value was 1.

F1 score for Training curve consistently reduced with increase in the number of iterations whereas Test score remained flat after 500 iterations. Model achieve f1 score of 0.69. Learning curve shows high training & cv score and small gap between training and cv curves which indicates that model has low variance and low bias.

**Comparison with Backpropagation:** Compare to model with backpropagation, Random Hill Climbing model took more time to train whereas testing time was same for both models. F1 score for Random Hill Climbing is slightly (0.01) lower than backpropagation model. In Random Hill Climbing model, although True Positive slightly lowered True negative improved.

## 6 Simulated Annealing

### 6.1 Model Tuning using Randomized Optimization Parameters

```python
model_stats = []
for mi in [100, 500, 1000]:
    for ma in [10, 50, 100]:
        for lr in [0.001, 0.01, 1]:
            
            nn_sa_clf = mlrose.NeuralNetwork(hidden_nodes = [30], activation = 'relu',
                                         algorithm = 'simulated_annealing', max_iters = mi,
                                         bias = True, is_classifier = True, learning_rate = lr, 
                                         max_attempts= ma, random_state = 11)

            t_start = time.time()*1000.0
            nn_sa_clf = nn_sa_clf.fit(Wine_X_train, Wine_y_train)
            t_stop = time.time()*1000.0
            traintime = t_stop - t_start

            t_start = time.time()*1000.0
            Wine_y_pred = cross_val_predict(nn_sa_clf, Wine_X_train, Wine_y_train, cv=cv, n_jobs=-1)
            t_stop = time.time()*1000.0
            testtime = t_stop - t_start

            prscore = precision_score(Wine_y_train, Wine_y_pred, average='binary')
            rcscore = recall_score(Wine_y_train, Wine_y_pred, average='binary')
            f1score = f1_score(Wine_y_train, Wine_y_pred, average='binary')

            # MI=100, MA=10, LR=0.001
            model_stats.append(['SA', ma, mi, lr, "{:.2f}".format(prscore), "{:.2f}".format(rcscore), "{:.2f}".format(f1score), traintime, testtime])
```

```python
dfstats_SA = pd.DataFrame(model_stats, columns=['Model', 'Max_Attempts', 'Max_Iterations', 'Learn_rate', 'PRScore', 'RCScore', 'F1Score', 'TrainTime', 'TestTime'])
dfstats_SA.sort_values(['F1Score', 'TestTime'], ascending=[False, True]).iloc[0]
```

```python
dfstats_SA[['PRScore','RCScore','F1Score']]=dfstats_SA[['PRScore','RCScore','F1Score']].astype(float)
fig, ax = plt.subplots(figsize=(15,7))
dfstats_SA.groupby(['Max_Iterations','Max_Attempts','Learn_rate']).max()['F1Score'].unstack().plot(ax=ax)

plt.title('SA: Fitness Vs Max_iter, Max_Attempts & Learn_rate')
plt.legend(loc='lower right', title="Learning Rate")
plt.xlabel('(Max Attempts, Max Iterations)')
plt.ylabel('Fitness')
plt.minorticks_on()
plt.grid(b=True, which='major', color='k', linestyle='-', alpha=0.1)
plt.grid(b=True, which='minor', color='r', linestyle='-', alpha=0.05)
plt.show()
```

```python
model_stats = []

expdecay = mlrose.ExpDecay(init_temp=1.0, exp_const=0.005, min_temp=0.001)
arithdecay = mlrose.ArithDecay(init_temp=1.0, decay=0.0001, min_temp=0.001)
geomdecay = mlrose.GeomDecay(init_temp=1.0, decay=0.99, min_temp=0.001)

for decay in [expdecay, arithdecay, geomdecay]:
            
    nn_sa_clf = mlrose.NeuralNetwork(hidden_nodes = [30], activation = 'relu',
                                 algorithm = 'simulated_annealing', max_iters = mi,
                                 bias = True, is_classifier = True, learning_rate = lr, 
                                 max_attempts= ma, schedule=decay, random_state = 11)

    t_start = time.time()*1000.0
    nn_sa_clf = nn_sa_clf.fit(Wine_X_train, Wine_y_train)
    t_stop = time.time()*1000.0
    traintime = t_stop - t_start

    t_start = time.time()*1000.0
    Wine_y_pred = cross_val_predict(nn_sa_clf, Wine_X_train, Wine_y_train, cv=cv, n_jobs=-1)
    t_stop = time.time()*1000.0
    testtime = t_stop - t_start

    prscore = precision_score(Wine_y_train, Wine_y_pred, average='binary')
    rcscore = recall_score(Wine_y_train, Wine_y_pred, average='binary')
    f1score = f1_score(Wine_y_train, Wine_y_pred, average='binary')

    # MI=100, MA=10, LR=0.001
    model_stats.append(['SA', type(decay).__name__, "{:.2f}".format(prscore), "{:.2f}".format(rcscore), "{:.2f}".format(f1score), traintime, testtime])
```

```python
dfstats_SA = pd.DataFrame(model_stats, columns=['Model', 'Schedule', 'PRScore', 'RCScore', 'F1Score', 'TrainTime', 'TestTime'])
dfstats_SA.sort_values(['F1Score', 'TestTime'], ascending=[False, True]).iloc[0]
```

**Model Tuning:** Hyperparameters (are hidden nodes and activation function) of the Neural Network tuned using backpropagation method were set to optimal values of 30 and 'relu'. After that Simulated Annealing algorithm was tuned and optimal values of Max Attempts, Max Iterations and Learning rate were identified as 100, 1000 and 1 respectively. With these optimal parameters model was further tuned on the Schedule parameter and optimal value for Schedule was geomdecay.

### 6.2 Bais-Variance Analysis

```python
nn_sa_clf = mlrose.NeuralNetwork(hidden_nodes = [30], activation = 'relu',
                                         algorithm = 'simulated_annealing', max_iters = 1000,
                                         bias = True, is_classifier = True, learning_rate = 1, 
                                         max_attempts= 10, schedule=geomdecay, random_state = 11)

#plot learning curve
utils.plot_learning_curve(nn_sa_clf, "SA - Learning Curve", Wine_X_train, Wine_y_train, cv=cv, n_jobs=-1, exp_score=0.8, train_sizes=np.linspace(0.05, 1.0, 10), ylim=[0.5,1])

t_start = time.time()*1000.0
nn_sa_clf = nn_sa_clf.fit(Wine_X_train, Wine_y_train)
t_stop = time.time()*1000.0
traintime = t_stop - t_start

t_start = time.time()*1000.0
Wine_y_pred = nn_sa_clf.predict(Wine_X_test)
t_stop = time.time()*1000.0
testtime = t_stop - t_start

cnf_matrix = confusion_matrix(Wine_y_test, Wine_y_pred)
np.set_printoptions(precision=2)

# Plot non-normalized confusion matrix
plt.figure()
utils.plot_confusion_matrix(cnf_matrix, classes=[0,1], title='SA (Test)')
 
modelid = "SA (Test)"
prscore = precision_score(Wine_y_test, Wine_y_pred, average='binary')
rcscore = recall_score(Wine_y_test, Wine_y_pred, average='binary')
f1score = f1_score(Wine_y_test, Wine_y_pred, average='binary')

print("Precision: "+str("{:.2f}".format(prscore))+", Recall: "+str("{:.2f}".format(rcscore))+", F1 Score: "+str("{:.2f}".format(f1score))+", Train Time: "+str("{:.2f}".format(traintime))+", Test Time: "+str("{:.2f}".format(testtime)))
```

**Bias-Variance Analysis:** Tuned model with optimal parameters was executed on the Training and Testing sets with 10-fold cross validation. Model achieve f1 score (test) of 0.69. Performance of Simulated Annealing model was much lower than that of Random Hill climbing mode. Learning curve shows high training & cv score and small gap between training and cv curves which indicates that model has low variance and low bias.

**Comparison with Backpropagation:** Compare to model with backpropagation, Simulated Annealing model took 3x more time to train whereas testing time was same for both models. F1 score for Simulated Annealing substantially lowered than backpropagation model. In Simulated Annealing model, True Positive also substantially lowered and True negative slightly improved.

### 7.1 Model Tuning using Randomized Optimization Parameters

## 8 Conclusion

As expected, Back propagation was much faster as compared to all optimization algorithms because back propagation adjusts the weights in the right direction based on actual output whereas optimization algorithms relies of random changes and do not leverage available information.  
Backpropagation tries to minimize error by learning the error and correcting it in each iteration. Neural network with back propagation is adaptive to the error and hence faster than GA.  
Performance of Genetic algorithm was better compared to all other methods. It takes advantage of random selection, mutation and cross over to avoid being caught in a local optimum. Performance of back propagation is lower mostly likely because it can get stuck in local optima.
