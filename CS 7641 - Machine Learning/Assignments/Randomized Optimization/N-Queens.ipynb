```python
import ro_ml_util as utl
import mlrose_hiive as mlrose_hiive
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
import pandas as pd

SEED = 42
```

```python
sz = 20
```

```python
problem = mlrose_hiive.generators.QueensGenerator.generate(seed=SEED, size=sz)
```

```python
a = problem.state
```

```python
print(a)
```

```python
board = np.full(shape=(sz,sz), fill_value=0)
```

```python
for i in range(len(a)):
    board[i][a[i]] = 1
```

```python
print(board)
df = pd.DataFrame(board, index=np.arange(0, board.shape[0]), columns=np.arange(0, board.shape[0]))
plt.close("all")
plt.figure(figsize=(10,7))
sn.heatmap(df, annot=True)
```

```python
rhc_best_params = utl.find_best_randomized_hill_climbing(problem=problem, max_attempts=[1000, 2000, 5000],
                                                     restarts=[0, 1, 3], seed=SEED)
```

```python
results = rhc_best_params[0]
board = np.full(shape=(sz, sz), fill_value=0)
for i in range(len(results)):
    board[i][results[i]] = 1
```

```python
print(board)
df = pd.DataFrame(board, index=np.arange(0, board.shape[0]), columns=np.arange(0, board.shape[0]))
plt.close("all")
plt.figure(figsize=(10,7))
sn.heatmap(df, annot=True)
```
