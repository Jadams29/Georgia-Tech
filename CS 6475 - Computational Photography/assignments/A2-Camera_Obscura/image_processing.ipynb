```python
import cv2
import numpy as np
import scipy as sp

import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("ButterFly.jpg", cv2.IMREAD_COLOR)
img_grayscale = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img); plt.axis("off"); plt.title("Butter Fly");
```

```python
horizontal_kernel = np.asarray([[0, 0],[-1, 1],[0, 0]])
vertical_kernel = np.asarray([[0, -1, 0], [0, 1, 0]])
plt.figure(figsize=(24, 4))

horizontal_test = cv2.filter2D(img_grayscale, -1, horizontal_kernel)
vertical_test = cv2.filter2D(img_grayscale, -1, vertical_kernel)

plt.subplot(131); plt.imshow(horizontal_test, cmap="gray"); plt.axis("off"); plt.title("Horizontal Test")
plt.subplot(132); plt.imshow(vertical_test, cmap="gray"); plt.axis("off"); plt.title("Vertical Test")
```

```python
combined_kernel = np.asarray([[0, -1, 0], [-1, 0, 1], [0, 1, 0]])
combined_test = cv2.filter2D(img_grayscale, -1, combined_kernel)

plt.figure(figsize=(24, 4))
plt.subplot(131); plt.imshow(combined_test, cmap="gray"); plt.axis("off"); plt.title("Combined Test")
```

```python
tiger_img = cv2.imread('Tiger.jpg', cv2.IMREAD_COLOR)
tiger_grayscale_img = cv2.cvtColor(tiger_img, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(24, 4))
plt.subplot(131); plt.imshow(tiger_img); plt.axis("off"); plt.title("Tiger Image")
plt.subplot(132); plt.imshow(tiger_grayscale_img, cmap='gray'); plt.axis("off"); plt.title("Tiger Grayscale Image")
```

```python
plt.figure(figsize=(24, 4))
box_blur_img = cv2.blur(tiger_grayscale_img, (3, 3))
gaussian_blur_img = cv2.GaussianBlur(tiger_grayscale_img, (3, 3), cv2.BORDER_DEFAULT)
plt.subplot(131); plt.imshow(box_blur_img, cmap='gray'); plt.axis("off"); plt.title("CV2 Box Blurred Tiger Image")
plt.subplot(132); plt.imshow(gaussian_blur_img, cmap='gray'); plt.axis("off"); plt.title("Gaussian Blur Tiger Image")
```

```python
box_blur_edge_img = cv2.filter2D(box_blur_img, -1, combined_kernel)
gaussian_blur_edge_img = cv2.filter2D(gaussian_blur_img, -1, combined_kernel)
edge_grayscale_img = cv2.filter2D(tiger_grayscale_img, -1, combined_kernel)

plt.figure(figsize=(24, 4))
plt.subplot(131); plt.imshow(box_blur_edge_img, cmap='gray'); plt.axis("off"); plt.title("CV2 Box Blurred Edge Tiger Image")
plt.subplot(132); plt.imshow(gaussian_blur_edge_img, cmap='gray'); plt.axis("off"); plt.title("Gaussian Blur Edge Tiger Image")
plt.subplot(133); plt.imshow(edge_grayscale_img, cmap='gray'); plt.axis("off"); plt.title("Grayscale Edge Tiger Image")
```

```python
sobel_x_kernel = np.asarray([[-1,0,1],[-2,0,2],[-1,0,1]])
sobel_y_kernel = np.asarray([[-1,-2,-1],[0,0,0],[1,2,1]])

sobel_x_img = cv2.filter2D(tiger_grayscale_img, -1, sobel_x_kernel)
sobel_y_img = cv2.filter2D(tiger_grayscale_img, -1, sobel_y_kernel)

plt.figure(figsize=(24, 4))
plt.subplot(131); plt.imshow(sobel_x_img, cmap='gray'); plt.axis("off"); plt.title("Sobel X Edge Image")
plt.subplot(132); plt.imshow(sobel_y_img, cmap='gray'); plt.axis("off"); plt.title("Sobel Y Edge Image")
```

```python
prewitt_x_kernel = np.asarray([[-1,0,1],[-1,0,1],[-1,0,1]])
prewitt_y_kernel = np.asarray([[-1,-1,-1],[0,0,0],[1,1,1]])

prewitt_x_img = cv2.filter2D(tiger_grayscale_img, -1, prewitt_x_kernel)
prewitt_y_img = cv2.filter2D(tiger_grayscale_img, -1, prewitt_y_kernel)

plt.figure(figsize=(24, 4))
plt.subplot(131); plt.imshow(prewitt_x_img, cmap='gray'); plt.axis("off"); plt.title("Prewitt X Edge Image")
plt.subplot(132); plt.imshow(prewitt_y_img, cmap='gray'); plt.axis("off"); plt.title("Prewitt Y Edge Image")
```

```python
roberts_x_kernel = np.asarray([[0, 1],[-1, 0]])
roberts_y_kernel = np.asarray([[1,0], [0, -1]])
print(tiger_grayscale_img.dtype)
print(roberts_x_kernel.dtype)

roberts_x_img = cv2.filter2D(tiger_grayscale_img, -1, roberts_x_kernel)
roberts_y_img = cv2.filter2D(tiger_grayscale_img, -1, roberts_y_kernel)

plt.figure(figsize=(24, 4))
plt.subplot(131); plt.imshow(roberts_x_img, cmap='gray'); plt.axis("off"); plt.title("Roberts X Edge Image")
plt.subplot(132); plt.imshow(roberts_y_img, cmap='gray'); plt.axis("off"); plt.title("Roberts Y Edge Image")
```

```python
test = np.full((3,3), 2, dtype=float)
kernel = np.asarray([[1,1,1], [1,1,1], [1,1,1]])


border_replicate = cv2.filter2D(test, -1, kernel, cv2.BORDER_REPLICATE)
border_reflect = cv2.filter2D(test, -1, kernel, cv2.BORDER_REFLECT)
border_wrap = cv2.filter2D(test, -1, kernel, cv2.BORDER_WRAP)

border_reflect_101 = cv2.filter2D(test, -1, kernel, cv2.BORDER_REFLECT_101)
border_isolate = cv2.filter2D(test, -1, kernel, cv2.BORDER_ISOLATED)
border_transparent = cv2.filter2D(test, -1, kernel, cv2.BORDER_TRANSPARENT)

print("\nborder_replicate\n",border_replicate)
print("\nborder_reflect\n",border_reflect)
print("\nborder_reflect_101\n",border_reflect_101)

print("\nborder_wrap\n",border_wrap)
print("\nborder_isolate\n",border_isolate)
print("\nborder_transparent\n",border_transparent)
```

```python
from scipy import signal

test = np.full((3,3), 2, dtype=float)
kernel = np.asarray([[1,1,1], [1,1,1], [1,1,1]])

same = signal.convolve2d(test, kernel, boundary='symm', mode='same')
valid = signal.convolve2d(test, kernel, boundary='symm', mode='valid')
full = signal.convolve2d(test, kernel, boundary='symm', mode='full')

print("\nSame Symm\n", same)
print("\nValid Symm\n", valid)
print("\nFull Symm\n", full)

same_wrap = signal.convolve2d(test, kernel, boundary='wrap', mode='same')
valid_wrap = signal.convolve2d(test, kernel, boundary='wrap', mode='valid')
full_wrap = signal.convolve2d(test, kernel, boundary='wrap', mode='full')

print("\nSame Wrap\n", same_wrap)
print("\nValid Wrap\n", valid_wrap)
print("\nFull Wrap\n", full_wrap)

same_fill = signal.convolve2d(test, kernel, boundary='fill', mode='same')
valid_fill = signal.convolve2d(test, kernel, boundary='fill', mode='valid')
full_fill = signal.convolve2d(test, kernel, boundary='fill', mode='full')

print("\nSame Fill\n", same_fill)
print("\nValid Fill\n", valid_fill)
print("\nFull Fill\n", full_fill)
```
