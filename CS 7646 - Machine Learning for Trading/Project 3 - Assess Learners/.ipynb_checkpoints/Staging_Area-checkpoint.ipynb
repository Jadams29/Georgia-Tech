```python
def run_dt_learner(train_x, train_y, test_x=None, leaf_size=1, fname=None):
    # Learner with leaf_size 1
    learner1 = dtl.DTLearner(leaf_size=leaf_size, verbose=False)
    print(f"Decision Tree: {fname}\n\tLeaf Size: {leaf_size}")
    print(f"\tLearner Leaf Size: {learner1.leaf_size}")
    print("\tAuthor Should be: jadams334")
    print(f"\tLearner Author: {learner1.author()}")
    learner1.add_evidence(train_x, train_y)
    y_pred_1_in_sample = learner1.query(train_x)
    if test_x is None:
        test_x = train_x
    y_pred_1_out_sample = learner1.query(test_x)
    return y_pred_1_in_sample, y_pred_1_out_sample

def run_rt_learner(train_x, train_y, test_x, leaf_size=1):
    # Learner with leaf_size 1
    learner1 = rtl.RTLearner(leaf_size=leaf_size, verbose=False)
    print(f"Random Tree:\n\tLeaf Size: {leaf_size}")
    print(f"\tLearner Leaf Size: {learner1.leaf_size}")
    print("\tAuthor Should be: jadams334")
    print(f"\tLearner Author: {learner1.author()}")
    learner1.add_evidence(train_x, train_y)
    y_pred_1_out = learner1.query(test_x)
    y_pred_1_in = learner1.query(train_x)
    return y_pred_1_in, y_pred_1_out

def run_bag_learner(train_x, train_y, test_x, learner, bags=1, learner_name="", **kwargs):
    # Learner with 1 bag
    learner1 = bg.BagLearner(learner=learner, bags=bags, kwargs=kwargs)
    learner1.add_evidence(train_x, train_y)
    y_pred_1 = learner1.query(test_x)
    print(f"Bag Learner:\n\tNumber of Bags: {bags}")
    print(f"\tLearner type should be: {learner_name}")
    print(f"\tCurrent Learner: {learner.__name__}")
    print(f"\tLearner Number of Bags: {learner1.bags}")
    print("\tAuthor Should be: jadams334")
    print(f"\tLearner Author: {learner1.author()}")
    return y_pred_1

def run_insane_learner(train_x, train_y, test_x,):
    learner = il.InsaneLearner(verbose=False)
    learner.add_evidence(train_x, train_y)
    y_pred = learner.query(test_x)
    return y_pred

def run_linear_learner(train_x, train_y, test_x,):
    return

def check_results(bag=False):
    points = 0

    return points
```
